Write a Pandas program to filter all records where the average consumption of beverages per person from .5 to 2.50 in world alcohol consumption dataset

```python
df = pd.read_csv("world_alcohol - world_alcohol.csv")
df.head()
```

```python
df_2 = df.groupby("Country")[["Display Value"]].mean().sort_values(by = "Display Value", ascending = False)
df_2[(df_2["Display Value"] >= 0.50) & (df_2["Display Value"] <= 2.50)]
```
